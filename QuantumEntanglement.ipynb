```python
from qiskit import *
import numpy as np
from qiskit.providers.ibmq import least_busy
from qiskit.visualization import plot_histogram
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.measure([0,1],[0,1])
qc.draw('mpl')
```

```python
backend = BasicAer.get_backend('qasm_simulator')
plot_histogram(execute(qc,backend,shots = 4096).result().get_counts())
```
